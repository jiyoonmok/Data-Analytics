# 서울시 구별 CCTV 설치 현황 분석

### 1. 데이터 준비

+ http://opengov.seoul.go.kr/data/11124500
    - CCTV 현황

+ https://data.seoul.go.kr/dataList/419/S/2/datasetView.do
    - 서울시 인구

```python
##### CCTV 자료

cctv_seoul = pd.read_csv("data/seoul/cctv.csv")
cctv_seoul.head()
```

```python
##### 인구 자료

pop_seoul = pd.read_excel("data/seoul/Report.xls", usecols="B, D, G, J, N", header=2) 
# 사용할 엑셀 열만 가져올 수 있음 # 두번째 열을 열 이름으로
pop_seoul.head()
```

### 2. EDA & Preprocessing

#### 2-1. 컬럼 이름 변경

```python
##### CCTV 자료에서 '기관명' 이라는 컬럼을 '구별'로 변경

cctv_seoul.rename(columns={"기관명":"구별"},inplace=True)
cctv_seoul.head()
```

```python
##### 인구 수 자료에서 컬럼명을 각각 구별, 인구수, 한국인, 외국인, 고령자
pop_seoul.columns=["구별","인구수","한국인","외국인","고령자"]
pop_seoul.head()
```

#### 2-2. CCTV 현황 탐색

+ 어느 구가 가장 CCTV가 많이 설치되어 있는가? 또는 가장 적게 설치가 되어있는가?

```python
##### 단순히 구별로 CCTV의 많고 적음을 탐색

cctv_seoul.sort_values(by="소계",ascending=False).iloc[:,:2]
```

```python
##### CCTV 설치 증가율이 가장 높은 구와 가장 낮은 구 
##### CCTV의 설치 연도별 증가율을 구해서 "최근 증가율" 파생변수 추가
##### ((2014 + 2015 + 2016) / 2013년도 이전) * 100

cctv_seoul["최근증가율"] = ((cctv_seoul["2014년"] + cctv_seoul["2015년"] + cctv_seoul["2016년"]) / cctv_seoul["2013년도 이전"])*100
cctv_seoul.sort_values(by="최근증가율",ascending=False)[["구별","최근증가율"]].iloc[[0,1]]
```

#### 2-3. 인구수 탐색

- 어느 구가 가장 인구수가 많은가? 또는 적은가?

```python
##### 필요없는 행 삭제 (서울시 전체 합계 삭제) : del, drop()
pop_seoul.drop([0], inplace=True)
pop_seoul.head()
```

```python
##### 구별 데이터의 중복이나 결측치 등을 확인

print(len(pop_seoul["구별"].unique()))
print(pop_seoul["구별"].count())
# 한개의 결측치가 있음을 확인

pop_seoul[pop_seoul["구별"].isnull()]
```

```python
##### 결측치 삭제
pop_seoul.drop([26],inplace=True)
```

```python
pop_seoul[pop_seoul["구별"].isnull()]
```

```python
##### 단순히 인구수만 탐색 (전체, 외국인, 고령자)

print(pop_seoul.sort_values(by="인구수",ascending=False).head())
print(pop_seoul.sort_values(by="인구수").head())

print("----------------------------------------------------------")
print(pop_seoul.sort_values(by="외국인",ascending=False).head())
print(pop_seoul.sort_values(by="외국인").head())

print("----------------------------------------------------------")
print(pop_seoul.sort_values(by="고령자",ascending=False).head())
print(pop_seoul.sort_values(by="고령자").head())
```

```python
##### 전체 인구 대비 외국인비율, 고령자비율 파생변수

pop_seoul["외국인비율"] = pop_seoul["외국인"]/pop_seoul["인구수"] * 100
pop_seoul["고령자비율"] = pop_seoul["고령자"]/pop_seoul["인구수"] * 100

print(pop_seoul.sort_values(by="외국인비율",ascending=False).head(5))
print("-----------------------------------------------------------------------------------")
print(pop_seoul.sort_values(by="고령자비율",ascending=False).head(5))
```

#### 2-4. 인구대비 CCTV 설치 현황

```python
##### CCTV 데이터와 인구 수 데이터를 합치기

data_result = pd.merge(cctv_seoul, pop_seoul, on="구별")
data_result.head()
```

```python
##### 필요없는 컬럼 삭제 

del data_result["2013년도 이전"]
del data_result["2014년"]
del data_result["2015년"]
del data_result["2016년"]
```

```python
data_result.head()
```

```python
##### 구별 컬럼을 인덱스로 변경
data_result.set_index("구별",inplace=True)
data_result.head()
```

```python
##### CCTV의 갯수(소계)와 인구 데이터 중 어느 인구가 관계가 있는지를 확인
##### np.corrcoef() : -1 ~ 1

print(np.corrcoef(data_result["고령자비율"], data_result["소계"]))
print(np.corrcoef(data_result["외국인비율"], data_result["소계"]))
print(np.corrcoef(data_result["인구수"], data_result["소계"]))
```

```python
##### 중간 체크 포인트

data_result.to_csv("data/seoul/cctv_result.csv", sep=",", encoding="utf-8")
```

```python
##### 파일 불러오기

data_result = pd.read_csv("data/seoul/cctv_result.csv", index_col="구별")
data_result
```

### 3. Visualization 

#### 3-1. CCTV가 많이 설치된 구 확인

```python
data_result["소계"].sort_values().plot(kind="barh", figsize=(10,5), grid=True, )
```

#### 3-2. 인구 대비 CCTV 개수 확인

```python
##### "CCTV비율" 파생 변수 추가 (소계 /인구수 *100)

data_result["CCTV비율"] = data_result["소계"] / data_result["인구수"] * 100
data_result.head()
```

```python
data_result["CCTV비율"].sort_values().plot(kind="barh", figsize=(10,5), grid=True)
```

```python
##### CCTV 개수과 인구수가 서로 관계가 있는지 산포도 확인

plt.figure(figsize=(6, 6))
plt.scatter(data_result["인구수"],data_result["소계"],s=60)
plt.xlabel("인구 수")
plt.ylabel("CCTV")
plt.grid()
```

```python
##### 회귀선 작성(y = wx + b)

fp1 = np.polyfit(data_result["인구수"], data_result["소계"], 1)
print(fp1) #기울기 , 절편
 
# 회귀식으로 만들어줌
f1 = np.poly1d(fp1)
print(f1)

# x값
fx = np.linspace(100000, 700000, 100)

plt.figure(figsize=(6, 6))
plt.scatter(data_result["인구수"],data_result["소계"],s=60)
plt.plot(fx, f1(fx), lw=3, color="g", ls="dashed")
plt.xlabel("인구 수")
plt.ylabel("CCTV")
plt.grid()
```

```python
##### 각 데이터의 오차를 구해서 "오차"라는 파생변수 추가

data_result["오차"] = np.abs(data_result["소계"] - f1(data_result["인구수"]))
```

```python
df_sort = data_result.sort_values(by="오차",ascending=False)
df_sort
```

```python
# 오차에 따라 색을 다르게 설정
plt.figure(figsize=(6, 6))
plt.scatter(data_result["인구수"],data_result["소계"],s=60, c=data_result["오차"]) # 오차 설정
plt.plot(fx, f1(fx), lw=3, color="g", ls="dashed")
plt.xlabel("인구 수")
plt.ylabel("CCTV")
plt.grid()
plt.colorbar() # 색깔에 대한 범례

for n in range(5):
    plt.text(df_sort["인구수"][n], df_sort["소계"][n], df_sort.index[n], fontsize=10)
```

```python
##### 최종 결과물 파일로 저장

df_sort.to_csv("data/seoul/cctv_result.csv", sep=",", encoding="utf-8")
```
