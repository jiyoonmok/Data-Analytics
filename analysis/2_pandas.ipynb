```python
#### 인덱스를 자유롭게 지정

s = pd.Series([9904312, 3448737, 2890451, 2466052], index = ["서울","부산","인천","대구"])
print(s)

print("-----------------------------------------------------------------")

print(s.values, type(s.values))
print(s.index)

print("-----------------------------------------------------------------")

# 인덱스에 이름 지정
s.index.name = "도시"
print(s)

print("-----------------------------------------------------------------")

# Series 전체에 이름 지정
s.name = "인구"
print(s)
```

```python
##### 벡터화 연산

s1 = s / 1000000
print(s1)
```

```python
##### 인덱싱

print(s[1], s["부산"])

print("------------------------------------------------")
print(s[1:3])

print("------------------------------------------------")
print(s["부산":"대구"]) # 대구까지 포함시킴

print("------------------------------------------------")
print(s[[0, 3, 1]]) # 필요한 데이터만 갖고오기

print("------------------------------------------------")
print(s[["서울","대구","부산"]])

print("------------------------------------------------")
# facny index
print(s[(s>250e4) & (s<500e4)])

print("------------------------------------------------")
print(s.부산, "," , s.서울)
```

```python
##### Series 와 dict

print("서울" in s)
print("대전" in s)

print("------------------------------------------------")

print(list(s.items()))

print("------------------------------------------------")

s2 = pd.Series({"서울":9631482, "부산":3448737, "인천":2632035, "대전":1490158})
print(s2)
```

```python
##### 인덱스 기반 연산
print(s)
print("------------------------------------------------")
print(s2)
print("------------------------------------------------")

# 인덱스 끼리 매칭되는것만 연산 아닌것은 null값
result = s-s2
print(result)

print("------------------------------------------------")
# 인덱스에 상관없이 순서대로 연산
print(s.values - s2.values)
```

```python
##### 결측치 제거

print(result)
print("------------------------------------------------")
print(result.notnull())
print("------------------------------------------------")
print(result[result.notnull()])

# s와 s2 데이터를 이용해서 인구증가율 : (끝연도 - 시작연도) / 시작연도*100
print("------------------------------------------------")
result2 = (s2-s) / s*100
print(result2)
print("------------------------------------------------")
result2 = result2[result2.notnull()]
print(result2)
```

```python
##### 데이터를 수정, 삭제, 갱신

# 수정
result2["부산"] = 1.63
print(result2)
print("------------------------------------------------")

# 갱신
result2["대구"] = 1.41
print(result2)
print("------------------------------------------------")

# 삭제
del result2["서울"]
print(result2)
```

### (5) apply()

+ 행이나 열 단위로 더 복잡한 처리를 하고자 할 때 사용
+ 인수로 행 또는 열을 받는 함수를 apply()의 인자로 넣으면 각 열(또는 행)을 반복하여 그 함수에 적용

```python
df = pd.DataFrame({
    "A" : [1, 3, 4, 3, 4],
    "B" : [2, 3, 1, 2, 3],
    "C" : [1, 5, 2, 4, 4]
})
df
```

```python
##### 각 열(행)별로 최대값에 최소값을 뺀 값을 구하고자 한다. 

def diff(x):
    return x.max() - x.min()

print(diff(df["A"]))
print(diff(df["B"]))
print(diff(df["C"]))

print("----------------------------------")

# 열 기준
print(df.apply(diff, axis=0))
# 행 기준
print(df.apply(diff, axis=1))
```

```python
##### 람다 함수
df.apply(lambda x:x.max()-x.min(), axis=0)
```

```python
##### 각 열에 대해 어떤 값이 얼마나 사용되었는지 조회

print(df["A"].value_counts())
print(df["B"].value_counts())
print(df["C"].value_counts())

print("----------------------------------")

df.apply(pd.value_counts, axis=0)
# df.apply(lambda x:x.value_counts(),axis=0)
```

```python
##### 다중 컬럼과 다중 인덱스

df3 = pd.DataFrame(np.round(np.random.rand(6, 4), 2),
                columns=[["A","A","B","B"],["C","D","C","D"]],
                  index = [["M","M","M","F","F","F"], ["id_" + str(i+1) for i in range(3)]* 2])

df3.columns.names = ["Cidx1","Cidx2"]
df3.index.names = ["Ridx1","Ridx2"]
df3
```

```python
"""
loc 인덱서
Ridx2
id_1   0.14
"""
print(df3.loc["M"]["A"]["C"][:1])

print("----------------------------")


"""
기본 인덱서
0.14
"""
print(df3[("A","C")][("M","id_1")])
# print(df3[("A","C")][0])

print("----------------------------")

"""
iloc 인덱서
0.14
"""
print(df3.iloc[0,0])
# print(df3.iloc[0][0])

print("----------------------------")

"""
iloc 인덱서
Ridx1 Ridx2
M     id_1    0.14
Name: (A, C), dtype:float64
"""
print(df3.iloc[:1,0])
```

```python
##### 행 인덱스와 열 인덱스를 교환 : stack()

print(df3.T)

print("-------------------------------------")

# Cidx1 열 인덱스를 행 인덱스로 바꿔줌
df3.stack("Cidx1")
df3.stack(0)

# Cidx2 열 인덱스를 행 인덱스로 바꿔줌
df3.stack("Cidx2")
df3.stack(1)

# Ridx1 행 인덱스를 열 인덱스로 바꿔줌
df3.unstack("Ridx1")
df3.unstack(0)

# Ridx2 행 인덱스를 열 인덱스로 바꿔줌
df3.unstack("Ridx2")
df3.unstack(1)
```

```python
##### 순서교환 : swaplevel(i, j, axis=0)

# Ridx1과 Ridx2 순서 교환
df4 = df3.swaplevel("Ridx1","Ridx2")
df4

# Cidx1과 Cidx2 순서 교환
df5 = df3.swaplevel("Cidx1","Cidx2",axis=1)
df5
```

```python
##### apply(), agg(), transform()

import seaborn as sns
iris = sns.load_dataset("iris")
iris.head()
```

```python
# 각 품종별로 가장 큰 값과 가장 작은 값의 비율을 조회

def peek_to_peek_ratio(x):
    return x.max()/x.min()

#######################################################

g = iris.groupby("species")

print(g.apply(peek_to_peek_ratio))
print("-------------------------------------------------")
print(g.agg(peek_to_peek_ratio))
print("-------------------------------------------------")
print(g.transform(peek_to_peek_ratio))
```

```python
import seaborn as sns

tips = sns.load_dataset("tips")
tips.head()
```

```python
##### 식사 대금과 팁의 비율 (tip / total_bill)을 나타내는 tip_pct라는 파생변수 추가

tips["tip_pct"] = (tips["tip"] / tips["total_bill"]).round(2)
tips.head()
```

```python
##### 성별로 인원 수 파악
tips["sex"].value_counts()
tips.groupby("sex").count()
tips.groupby("sex").size()
```

```python
##### 위의 두 정보를 하나의 테이블로 tip_pct 조회

tips.pivot_table(index="sex",columns="smoker",values="tip_pct")
# tips.pivot_table(index=["sex","smoker"], aggfunc = "mean", values=["size","tip_pct"])
tips.pivot_table("tip_pct", index=["sex","smoker"], aggfunc="count", margins=True)
```

```python
##### 성별에 따른 평균 팁 비율 (groupby(), pivot_table())
tips.groupby("sex").mean()[["tip_pct"]]
tips.pivot_table(values="tip_pct",index="sex",aggfunc="mean")
```

```python
##### 팁의 비율이 요일과 점심/저녁 여부, 인원수에 어떤 영향을 받는지 살펴볼 수 있게 조회
tips.pivot_table("tip_pct",index=["day","time"],columns="size").dropna()
tips.groupby(["day","time","size"]).mean()[["tip_pct"]].dropna()
```

```python
##### 성별, 흡연유무별로 가장 많은 팁과 가장 적은 팁의 차이
#tips.groupby(["sex","smoker"])[["tip"]].max() - tips.groupby(["sex","smoker"])[["tip"]].min() 

def peak_to_peak(x):
    return x.max() - x.min()

tips.groupby(["sex","smoker"])[["tip"]].apply(peak_to_peak)
tips.groupby(["sex","smoker"])[["tip"]].agg(peak_to_peak)

```

```python
##### shift 연산

np.random.seed(0)

ts = pd.Series(np.random.randn(4), index=pd.date_range("2021-1-1", periods=4, freq="M"))
print(ts)
print("-------------------------------------------")

print(ts.shift(1))
print("-------------------------------------------")

print(ts.shift(-1))
print("-------------------------------------------")

print(ts.shift(1,freq="M")) # 인덱스가 한달 이동
print("-------------------------------------------")

print(ts.shift(1, freq="W"))
```

```python
##### resampling : 시간 간격을 다시 재조정
# 시간 구간이 작아지면 데이터양이 증가 : 업 샘플링
# 시간 구간이 커지면 데이터양이 감소 : 다운 샘플링

ts = pd.Series(np.random.randn(100), index=pd.date_range("2021-1-1",periods=100)) # freq="D" 생략
ts.tail(10)

# 다운 샘플링 : 데이터가 그룹으로 묶이기 때문에 집계연산을 해서 대표값을 구해야한다.
print(ts.resample("W").mean())
print("-------------------------------------------")

print(ts.resample("M").mean())
print("-------------------------------------------")

ts = pd.Series(np.random.randn(60), index=pd.date_range("2012-1-1",periods=60,freq="T"))
ts
print(ts.resample("10T").sum)
```

```python
##### up-sampling
# forward filling : 앞에서 나온 데이터를 그대로 사용
# backward filling : 뒤에서 나온 데이터를 앞에서 미리 사용

print(ts.head())
print("-------------------------------------------")

print(ts.resample("30S").ffill().head())
print("-------------------------------------------")
print(ts.resample("30S").bfill().head())
```
